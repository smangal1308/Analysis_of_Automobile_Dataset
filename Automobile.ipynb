```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data visualisation or EDA (Exploratory data analysis)

```python
auto_price = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/Automobile_price_data__Raw_.csv', sep = ',')

# csv: comma seperated values (data field is seperated by comma by default)
```

```python
auto_price
```

```python
auto_price.drop(['symboling','normalized-losses'],axis=1,inplace=True)
```

```python
# Replace '?' by np.nan
auto_price_df = auto_price.replace('?',np.nan)
```

```python
auto_price_df.isnull().sum()
```

```python
# Drop the **rows** with missing data for each column
auto_price_df.dropna(inplace = True)
```

```python
auto_price_df.isnull().sum()
```

```python
cols = ['bore','stroke','horsepower','peak-rpm','price']
auto_price_df[cols] = auto_price_df[cols].apply(pd.to_numeric)
```

#### Line Plot

```python
# Two types of Line Plots:

# Univariate analysis: Trend analysis
# Bivariate analysis: Relationship analysis
```

#### Univariate Analysis

```python
fig = plt.figure(figsize = (4,3))
plt.plot(auto_price_df['price'])
plt.show()
```

```python
fig = plt.figure(figsize = (4,3))
plt.plot(auto_price_df['price'].sort_values())
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure(figsize = (4,3))
plt.plot(auto_price_df['price'].sort_values().unique())
plt.show()
```

#### Bivariate analysis:

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure(figsize = (4,3))
plt.plot(auto_price_df.sort_values(by = 'price')['price'],auto_price_df.sort_values(by = 'price')['city-mpg'])
plt.show()
```

```python
# Interpretation of bivariate analysis:
# There is a inverse relation between price and city milage.
```

#### Bar Plot

```python
# Bar plots are used by categorical data to display counts.
# Univariate/Bivariate Analysis
```

```python
plt.style.use('seaborn-whitegrid')
auto_price_df[['price']].loc[10:75,].plot.bar(figsize = (15,7))
```

```python
auto_price_df['make'].value_counts()
```

```python
auto_price_df['make'].value_counts().plot.bar(figsize = (12,4), color = 'orangered')
```

#### Histogram

```python
# Continuous samples
# Univariate Analysis
# Bar graph is a graphical representation of categorical data whereas Histogram is a graphical representation of quantitative data.
```

```python
fig = plt.figure(figsize = (9,5))
# Bins = Number of intervals
plt.hist(auto_price_df['price'], color = 'crimson', bins = 20)
plt.title('Histogram Plot -  Auto Price', size  = 9)
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Interpretation: Positive skewness
```

#### Distribution Plot

```python
import warnings
warnings.filterwarnings('ignore')

fig = plt.figure(figsize = (10,5))
sns.distplot(auto_price_df['price'], bins = 20, color = 'deepskyblue')
plt.show()
```

#### Scatter Plot

```python
fig = plt.figure(figsize = (10,5))
# marker = shape of scatter plot 's' (square), *, 'd' (diamond), '^', '<', '>', 'o'
# alpha value = color transparency
sns.scatterplot(x = auto_price_df['city-mpg'], y = auto_price_df['price'], color = 'orangered', marker = 's', alpha = 0.3)
plt.show()
# Scatter plot = Plot to observe relationship between mpg and price. Both are inversely proportional to each other.
```

#### Pie Chart

```python
auto_price_df['make'].value_counts()
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (10,10))
plt.pie(auto_price_df['make'].value_counts(), labels= auto_price_df['make'].value_counts().index, autopct = '%0.0f%%')
plt.show()
```

#### Box Plot

```python
# It is used to show the distribution of continuous data
# It shows statistical distribution like- Quartiles (Q1, Q2, Q3) and Median (50% or Q2).
# This plot is useful in showing outliers.
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize = (10,2))
# Provide any continuous numerical data
# By default: the data is vertically oriented
# orient = 'h': means Horizontal orientation
sns.boxplot(auto_price_df['price'], orient = 'h')
```
